- 좌표변환 및 병합

```python
#pip install pyproj
#pip install folium
import pyproj
import folium
```

```python
my_df = pd.read_csv("data/서울시 종로구 일반음식점 인허가 정보.csv",
                 encoding='cp949',
                 usecols=['좌표정보(X)','좌표정보(Y)'])
```

```python
my_df.index=range(len(my_df))
my_df.tail()
```

```python
def project_array(coord, p1_type, p2_type):
    p1 = pyproj.Proj(init=p1_type)
    p2 = pyproj.Proj(init=p2_type)
    fx, fy = pyproj.transform(p1, p2, coord[:, 0], coord[:, 1])
    return np.dstack([fx, fy])[0]

coord = np.array(my_df)
coord
```

```python
p1_type = "epsg:2097"
p2_type = "epsg:4326"

result = project_array(coord, p1_type, p2_type)
result
```

```python
my_df['경도'] = result[:, 0]
my_df['위도'] = result[:, 1]
my_df
```

```python
df2 = pd.merge(df1,my_df,on = '좌표정보(X)',how= 'left').drop_duplicates()
```

```python
#이전 좌표계 삭제
df2 = df2.drop(['좌표정보(X)','좌표정보(Y)_x','좌표정보(Y)_y'],axis = 'columns')
```

```python
# 인덱스 설정및 결측치 처리
df2 = df2.set_index('관리번호')
df2['위도'] = df2['위도'].fillna(0)
df2['경도'] = df2['경도'].fillna(0)
```
